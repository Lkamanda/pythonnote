## 字符串的简单操作
- '+' 字符串拼接
- '*' 字符串的复制操作
- '[]' 字符串的索引操作 从[0]开始
- [: :] 字符串的切片操作
- [开始索引:结束索引:间隔值] (包含开始,不包含结束)


```python
str1 = 'lo'
str2 = 've'
print(str1+str2)
print('lo'+'ve')

print ('林'*10)

s = 'i like dog'
print(s[0])
print(s[-1])
print(s[-2])

print(s[2:7])
print(s)
```

```python
s = '0123456789'
print(s[2:7:2])
# 切整个字符串
print(s[:])

# 指定开始,不指定结束
print(s[2:])
# 指定结束,不指定开始
print(s[:4])
# 指定开始,不指定结束,并给出间隔值
print(s[2::2])
# 指定结束,不指定开始,并给出间隔值
print(s[:5:2])
```

```python
# -1 capitalize 首字母大写  返回的是字符串
s = '狗 i like dog'
print(id(s))
print(id(s.capitalize()))
print(s.capitalize())
```

```python
# -2 title() 将每个单词首字母变成大写   返回的是字符串
print(s.title())
```

```python
# -3 upper() 将所有字母变为大写  返回的是字符串
s1 = 'i am zjl'
a1= s1.upper()
print(a1)
print(s1)
```

```python
# -4 lower() 将所有字母为小写  返回字符串
print(a1)
a1.lower()
```

```python
# -5 swapcase() 大小写互换
s2 = 'I am Str'
print(s2.swapcase())
```

```python
# 计算字符串长度  返回整形  (不属于字符串的内置函数 )
len(s2)
```

```python
# -6 find() 查找指定字符串　查不到返回　－１  查到返回第一个查到的索引值
#    index()            查不到不报错
s = 'xiaaolinxiaolin'
s1=s.find('l',1,8)
print(s1)
s2 = s.index('l')
print(s2)
```

```python
s.index('l',6)
```

```python
s.find('m')
```

```python
# -7 count() 计算字符串出现次数    返回整形
s = 'zxyzxyzx'
print(s.count('z'))
print(s.count('m'))
print(s.count('z',1))
```

```python
# -8 startswith() 检测是否是以指定之母开头   返回布尔值

#     endswith() 检测是否是以指定字结束
s1 = 'i like dog'
print(s1.startswith('i'))
print(s1.startswith('l',2))  # 可以加入索引

print(s1.endswith('g'))
print(s1.endswith('0'))
print(s1.endswith('g'))
```

```python
# - 9                                    返回布尔值

# isupper() 检测所有字母是否是大写字母
s = 'zxy'
s1 = 'ZXy'
s2 = 'Z X Y'
print(s.isupper())
print(s1.isupper())
print(s2.isupper())

# islower() 检测所有字母是否是小写字母
print(s.islower(),s1.islower(),s2.islower())

# istitle() 检测是否已制定标题显示(每个单词首字母大写)
print(s1.istitle())
print(s2.istitle())
print(s.istitle)

```
